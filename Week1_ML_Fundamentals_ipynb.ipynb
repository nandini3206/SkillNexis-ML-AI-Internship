This notebook demonstrates data preprocessing on the Titanic dataset by exploring the data, handling missing values, encoding categorical variables, visualizing the data, and preparing it for machine learning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
```

## Load the Dataset

```python
df = pd.read_csv("Titanic-Dataset.csv")

print("Dataset Loaded Successfully!")
```

## Assignment 1 : Explore the Dataset

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Analyze Missing Values

```python
df.isnull().sum()
```

## Assignment 2 : Handle Missing Values

```python
# Mean Imputation (Demonstration)

mean_age = df["Age"].mean()

print("Mean Age :", mean_age)
```

```python
# Median Imputation

df["Age"] = df["Age"].fillna(df["Age"].median())

# Mode for categorical column

df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Fill Cabin

df["Cabin"] = df["Cabin"].fillna("Unknown")

df.isnull().sum()
```

## Assignment 3 : Encode Categorical Features

```python
label_encoder = LabelEncoder()

df["Sex"] = label_encoder.fit_transform(df["Sex"])

df.head()
```

```python
df = pd.get_dummies(df, columns=["Embarked"], drop_first=True)

df.head()
```

## Mini Project : Data Visualization

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"], bins=30, kde=True)

plt.title("Age Distribution")

plt.xlabel("Age")

plt.ylabel("Number of Passengers")

plt.show()
```

## Save Cleaned Dataset

```python
df.to_csv("cleaned_titanic.csv", index=False)

print("Cleaned dataset saved successfully!")
```

## Conclusion

The Titanic dataset was successfully explored and preprocessed. Missing values were handled, categorical variables were encoded, the age distribution was visualized, and the cleaned dataset was saved for future machine learning tasks.
